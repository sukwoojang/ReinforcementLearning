# Dummy Q-Learning

* Frozen Lake : Even if you know the way, ask.

## Q-function (state-action value function)

* input을 통해 가장 높은 점수를 주는 action을 취함 -> $ argmaxQ(s1,a) $(a값) & $ maxQ(s1,a) $(Q값)
* Optimal policy with Q를 선정함

## Finding, Learning Q

* Assume(believe) Q in s` exists
* My condition:
    * i am in s
    * when i do action a, i'll go to s'
    * when i do action a, i'll get reward r
    * Q in s', Q(s',a') exist
* How can we express Q(s,a) using Q(s',a')?
    * Q(s,a) = r + maxQ(s',a')
    


```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from gym.envs.registration import register
import random as pr

def rargmax(vector):
    m = np.amax(vector)
    indices = np.nonzero(vector == m)[0]
    return pr.choice(indices)

register(
    id = 'FrozenLake-v3',
    entry_point = 'gym.envs.toy_text:FrozenLakeEnv',
    kwargs={'map_name' : '4x4',
            'is_slippery' : False}
)
env = gym.make('FrozenLake-v3') # environment 생성

Q = np.zeros([env.observation_space.n, env.action_space.n]) # 16개의 S, 4개의 A
num_episodes = 2000 # 2000번 정도의 루프 학습

rList = []

for i in range(num_episodes):
    state = env.reset() # 환경 초기화
    rAll = 0
    done = False
    
    while not done:
        action = rargmax(Q[state, :]) 
        # randomargmax -> 어떤 액션 하나를 취함 초기에는 0
        
        new_state, reward, done, _ = env.step(action) # 액션을 실행 그에 관한 정보를 돌려줌
        
        Q[state, action] = reward + np.max(Q[new_state, :])
        
        rAll += reward
        state = new_state
    
    rList.append(rAll)
```

```python
print('Success rate : ' + str(sum(rList)/num_episodes))
print('L D R U')
print(Q)
plt.bar(range(len(rList)), rList, color = 'blue')
plt.show()
```
